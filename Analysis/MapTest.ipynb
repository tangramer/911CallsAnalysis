```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# import dataset1
file1 = "Seattle_Real_Time_Fire_911_Calls.csv"
data = pd.read_csv(file1, nrows=5000) # nrows is the number of data lines to read
data
```

```python
lon_mean = (data['Longitude'].max()+data['Longitude'].min())/2

lat_mean = (data["Latitude"].max()+data["Latitude"].min())/2
```

```python
long_mean
```

```python
lat_mean = 47.608013
lon_mean = -122.335167
```

```python
conditions = [
    (data['Longitude'] >= lon_mean) & (data['Latitude'] >= lat_mean),
    (data['Longitude'] < lon_mean) & (data['Latitude'] >= lat_mean),
    (data['Longitude'] >= lon_mean) & (data['Latitude'] < lat_mean),
    (data['Longitude'] < lon_mean) & (data['Latitude'] < lat_mean)]
choices = ['NE', 'NW', 'SE', 'SW']
data['Zone'] = np.select(conditions, choices)
data
```

```python
import plotly.express as px
fig = px.scatter_mapbox(data, lat="Latitude",lon="Longitude", color = "Zone")
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

```python
data['Year'] = pd.DatetimeIndex(data['Datetime']).year
data['Month'] = pd.DatetimeIndex(data['Datetime']).month
# Count calls in May 2019
total_rows = data.loc[(data['Month'] == 5)&(data['Year'] == 2019)].shape[0]
total_rows
```

```python
# Call types in May, 2019
pd.set_option('display.max_rows', None)
data.loc[(data['Month'] == 5)&(data['Year'] == 2019)].Type.value_counts()
```

```python
medic = data['Type'] == "Medic Response"
car_fire = data['Type'] == "MVI - Motor Vehicle Incident"
brush_fire = data['Type'] == "Aid Response"
auto_fire = data['Type'] == "Fire in Single Family Res"
year2019 = data['Year'] == 2019
test=data[(auto_fire | car_fire | brush_fire | medic) & year2019]
test
```

```python
filtered=test.sort_values(by=['Month'])
filtered
```

```python
df = data.loc[(data['Month'] == 5)&(data['Year'] == 2019)]
df
```

```python
df = data.loc[(data['Month'] == 5)&(data['Year'] == 2019)&(data['Type'] == 'Aid Response')]
df
```

```python
mean_lat = df.groupby(['Type']).Latitude.mean()
mean_lon = df.groupby(['Type']).Longitude.mean()
std_lat = df.groupby(['Type']).Latitude.std()
std_lon = df.groupby(['Type']).Longitude.std()
size = np.sqrt(np.square(std_lat)+np.square(std_lon))

```

```python
test = df.groupby(['Type'])[['Latitude', 'Longitude']].describe()
test = test.reset_index()
```

```python
test['Latitude']['count']
```

```python
test.drop(columns=['B', 'C'])
test['Size'] = np.sqrt(np.square(test['Latitude']['std'])+np.square(test['Longitude']['std']))
test
```

```python
df_filtered = test[test['Latitude']['count'] >= 100] 
df_filtered
```

```python
df_plot = pd.DataFrame(columns=['Type', 'Latitude', 'Longitude', 'Size'])
df_plot["Type"] = df_filtered['Type']
df_plot["Longitude"] = df_filtered['Longitude']['mean']
df_plot["Latitude"] = df_filtered['Latitude']['mean']
df_plot["Size"] = np.square(df_filtered['Size'])*1000
df_plot
```

```python
import plotly.express as px
fig = px.scatter_mapbox(df_plot, lat='Latitude', lon='Longitude', size = "Size", color = "Type")
#fig = px.scatter_mapbox(df, lat=["Latitude"]['mean'],lon=["Longitude"]['mean'], color = "Type",animation_frame="Month", size='Size',animation_group = "Type")
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```
